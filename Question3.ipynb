```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
main_df = pd.read_csv('arrhythmia.csv', header=None)
main_df.replace('?', np.nan, inplace=True)
```

```python
knn_imputer=KNNImputer(n_neighbors=5)
main_df = pd.DataFrame(knn_imputer.fit_transform(main_df), columns=main_df.columns)
main_df[279]=main_df[279].astype('int')
main_df.iloc[0:5,0:11]
```

```python
standard = StandardScaler()
main_df2 = pd.DataFrame(standard.fit_transform(main_df.iloc[:,:-1]), columns=range(279))
main_df2[279] = main_df[279]
main_df = main_df2
main_df
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(main_df.iloc[:,0:-1], main_df[279],test_size=0.2)
```

```python
knn_model_1 = KNeighborsClassifier(n_neighbors=1, metric='euclidean')
knn_model_30 = KNeighborsClassifier(n_neighbors=30, metric='euclidean')
knn_model_1.fit(X_train,Y_train)
knn_model_30.fit(X_train,Y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
Y_knn_predicted_1 = knn_model_1.predict(X_test)
Y_knn_predicted_30 = knn_model_30.predict(X_test)
```

```python
print(confusion_matrix(Y_knn_predicted_1, Y_test),'\n')
print(confusion_matrix(Y_knn_predicted_30, Y_test))
unique, counts = np.unique(Y_test, return_counts=True)
print('\n',dict(zip(unique, counts)))
pd.DataFrame(confusion_matrix(Y_knn_predicted_30, Y_test))
```

```python
print(classification_report(Y_knn_predicted_1, Y_test),'\n')
print(classification_report(Y_knn_predicted_30, Y_test))
```

```python
Y_train_pred_1=knn_model_1.predict(X_train)
Y_train_pred_30=knn_model_30.predict(X_train)
```

```python
print(confusion_matrix(Y_train_pred_1, Y_train),'\n')
print(confusion_matrix(Y_train_pred_30, Y_train))
unique, counts = np.unique(Y_train, return_counts=True)
dict(zip(unique, counts))
```

```python
print(classification_report(Y_train_pred_1, Y_train),'\n')
print(classification_report(Y_train_pred_30, Y_train))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'n_neighbors':list(range(1,31)),'metric':['euclidean','manhattan']}
```

```python
grid = GridSearchCV(knn_model_1,param_grid, verbose=3, cv=10, scoring='recall_weighted')
grid.fit(X_train,Y_train)
```

```python
grid.best_estimator_
```

```python
Y_knn_grid_prediction = grid.predict(X_test)
Y_knn_grid_prediction
```

```python
print(confusion_matrix(Y_knn_grid_prediction, Y_test),'\n')
print(classification_report(Y_knn_grid_prediction, Y_test))
pd.DataFrame(confusion_matrix(Y_knn_grid_prediction, Y_test))
```
