```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler,Normalizer,StandardScaler
```

## Question1

```python
main_df = pd.read_csv('Data.csv', skiprows=2)
main_df.drop(['1'], axis=1, inplace=True)
main_df
```

```python
S_Imputer = SimpleImputer()
df1= main_df.iloc[:,2:5]
df1 = pd.DataFrame(S_Imputer.fit_transform(df1), columns=['Population growth','Total population','Area (sq. km)'])
```

```python
knn_Imputer = KNNImputer(n_neighbors=5)
df1 = main_df.iloc[:,2:5]
df1 = pd.DataFrame(knn_Imputer.fit_transform(df1), columns=['Population growth','Total population','Area (sq. km)'])
# Update main_df
main_df.iloc[:,2:5] = df1
df1
```

```python
label=LabelEncoder()
df2 = main_df.iloc[:,:]
df2['CountryCode'] = label.fit_transform(df2['CountryCode'])
df2['International Visitors'] = label.fit_transform(df2['International Visitors'])
# Update main_df
main_df = df2
main_df
```

```python
df3 = main_df.iloc[:,:]
min_max_T = MinMaxScaler()
pd.DataFrame(min_max_T.fit_transform(df3[['Population growth','Total population','Coronavirus Cases', 'Area (sq. km)']]), 
             columns=['Population growth','Total population','Coronavirus Cases', 'Area (sq. km)'])
```

```python
normal_T = Normalizer()
pd.DataFrame(normal_T.fit_transform(df3[['Population growth','Total population','Coronavirus Cases', 'Area (sq. km)']]), 
             columns=['Population growth','Total population','Coronavirus Cases', 'Area (sq. km)'])
```

```python
standard_T = StandardScaler()
pd.DataFrame(standard_T.fit_transform(df3[['Population growth','Total population','Coronavirus Cases', 'Area (sq. km)']]), 
             columns=['Population growth','Total population','Coronavirus Cases', 'Area (sq. km)'])
```

```python
Min_corona = np.min(main_df['Coronavirus Cases'])
Max_corona = np.max(main_df['Coronavirus Cases'])
array_T = min_max_T.fit_transform(main_df[['Population growth','Total population', 'Area (sq. km)','Coronavirus Cases']])
array_main_df = main_df.to_numpy()
array_main_df[:,2:5] = array_T[:,0:3]
array_main_df[:,6] = array_T[:,3]
array_main_df
#Update main_df 
main_df = pd.DataFrame(array_main_df, columns=main_df.columns)
main_df['Population growth'] = main_df['Population growth'].astype('float')
main_df['Total population'] = main_df['Total population'].astype('float')
main_df['Area (sq. km)'] = main_df['Area (sq. km)'].astype('float')
main_df['Coronavirus Cases'] = main_df['Coronavirus Cases'].astype('float')
main_df
```

```python
df4 = main_df
Q1 = df4[['Population growth','Total population','Area (sq. km)']].quantile(0.25)
Q3 = df4[['Population growth','Total population','Area (sq. km)']].quantile(0.75)
IQR = Q3 - Q1
df4[((df4[['Population growth','Total population','Area (sq. km)']] > Q3 + IQR*1.5) | 
     (df4[['Population growth','Total population','Area (sq. km)']] < Q1 - IQR*1.5))]
# Update main_df
main_df = df4[((df4[['Population growth','Total population','Area (sq. km)']] <= Q3 + IQR*1.5) & 
     (df4[['Population growth','Total population','Area (sq. km)']] >= Q1 - IQR*1.5)).all(axis=1)]
main_df
```

```python
import statsmodels.formula.api as smf
from statsmodels.tools.eval_measures import rmse
```

```python
lm = smf.ols('Q("Coronavirus Cases") ~ Q("Population growth")+Q("Total population")+Q("Area (sq. km)")',data=main_df).fit()
lm.summary()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = main_df.iloc[:, 2:5].values
Y = main_df.iloc[:, 6].values
lm = LinearRegression()
lm.fit(X,Y)
lm.coef_
```

```python
Predicted = lm.predict(main_df[['Population growth','Total population','Area (sq. km)']])
Predicted = Predicted.apply(lambda x: x*(Max_corona-Min_corona) + Min_corona)
main_corona = main_df['Coronavirus Cases'].apply(lambda x: x*(Max_corona-Min_corona)+ Min_corona)
RMSE = rmse(main_corona, Predicted)
pow(RMSE,2)
```
